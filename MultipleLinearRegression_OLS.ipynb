## Linear Regression 

```python
import numpy as np 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
```

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

```python
# defining the model
model = LinearRegression(fit_intercept = True, normalize = False)
```

```python
# first we need to fit the model
fit_model = model.fit(x,y)
# now we can test the results by calculating the R-squared
R_sq = fit_model.score(x,y)
print("R-squared is: ", R_sq)
print("The intercept, the average if all indep vb's 0:",fit_model.intercept_)
print("The coefficient, the impact of main indep ceteris paribus:",fit_model.coef_)
```

```python
# when performing prediction after fitting when using test (new data)
x_new = np.arange(5).reshape((-1,1))
y_pred = fit_model.predict(x_new)
print("preditcted y:", y_pred, sep = '\n')
```

# Multiple Linear Regression 

```python
x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
x_new = [[0, 2], [4, 2], [13, 1], [20, 3], [30, 10], [42, 12], [50, 30], [58, 32]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)
print(x,y)
```

```python
# if we wanna add conntant as column
x = sm.add_constant(x)
x_new = sm.add_constant(x_new)
print(x, x_new)
```

```python
model = LinearRegression()
fit_model = model.fit(x,y)
R_sq = fit_model.score(x,y)
print("R-squared is: ", R_sq)
print("The intercept:",fit_model.intercept_)
print("The coefficients:",fit_model.coef_)
```

```python
y_pred = fit_model.predict(x_new)
print(y_pred)
```

# OLS

```python
model = sm.OLS(y,x)
fit_model = model.fit()
print(fit_model.summary())
```

```python
print('R_squared:', fit_model.rsquared)
print('R_squared_adjusted, in case of MLR what matters:', fit_model.rsquared_adj)
```

```python
# prediction
y_pred = fit_model.predict(x_new)
print('predicted response:', y_pred, sep='\n')
```
